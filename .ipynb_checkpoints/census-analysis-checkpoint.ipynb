## Birthplaces

```python
basic_data['birthPlaceNormalized'].unique()
```

```python
birthplaces = basic_data.groupby('birthPlaceNormalized').agg('count')
birthplaces = birthplaces.drop(birthplaces.columns[1:], axis = 1).rename(index=str, columns={'age': 'Count'})
birthplaces = birthplaces.reset_index()
birthplaces
```

```python
us_birthplaces = [i for i in birthplaces['birthPlaceNormalized'] if 'United States' in i]
foreign_birthplaces = [i for i in birthplaces['birthPlaceNormalized'] if i not in us_birthplaces]
```

```python
print('US born Singhs', sum(birthplaces[birthplaces['birthPlaceNormalized'].isin(us_birthplaces)]['Count']))
print('Foreign born Singhs', len(basic_data.index) - sum(birthplaces[birthplaces['birthPlaceNormalized'].isin(us_birthplaces)]['Count']))
```

```python
us_birthplaces
```

```python
foreign_birthplaces
```

```python
birthplace_mapping = {'Southeastern US': ['Alabama, United States', 'Florida, United States', 'Georgia, United States',
                                     'Mississippi, United States', 'North Carolina, United States', 'Virginia, United States', 'Texas, United States'],
                     'Western US (Outside of California)': ['Arizona, United States', 'Colorado, United States',
                                                           'Idaho, United States', 'Montana, United States', 'Nevada, United States',
                                                           'New Mexico, United States', 'Oregon, United States', 'Utah, United States', 'Washington, United States'],
                     'California': ['California, United States'],
                     'Midwestern US': ['Illinois, United States', 'Indiana, United States', 'Iowa, United States', 'Kansas, United States', 'Michigan, United States',
                                      'Ohio, United States', 'Oklahoma, United States', 'Wisconsin, United States'],
                     'Northeastern US': ['Connecticut, United States', 'New Jersey, United States', 'New York, United States', 'Pennsylvania, United States', 'Rhode Island, United States'],
                     'Elsewhere in Americas': ['Barbados', 'Canada', 'British Guinea', 'Grenada', 'Jamaica', 'Mexico', 'South America', 'West Indies'],
                     'India': ['Punjab, India', 'India'],
                     'Other Asian Countries': ['Asia', 'China', 'Sri Lanka', 'Philippines'],
                     'Europe': ['England', 'Hungary', 'Russia', 'Spain']}
```

```python
def assign_birthplace(location):
    for p in birthplace_mapping:
        if location in birthplace_mapping[p]:
            return p
basic_data['birthRegion'] = basic_data['birthPlaceNormalized'].apply(assign_birthplace)
region_counts = basic_data.groupby('birthRegion').agg('count')
region_counts = region_counts.drop(region_counts.columns[1:], axis = 1).rename(index=str, columns={'age': 'Count'})
```

```python
list(region_counts.index)
```

```python
data = [go.Bar(
            x=list(region_counts.index),
            y=region_counts['Count']
    )]

py.iplot(data)
```

## Ages

```python
basic_data['age'].unique()
```

```python
def process_age(age_string):
    if age_string == '<1':
        return 0
    else:
        return int(age_string)
basic_data['age'] = basic_data['age'].apply(process_age)
```

```python
ages = [go.Histogram(x=basic_data['age'])]
py.iplot(ages)
```

```python
np.average(basic_data['age'])
```

```python
us_born = basic_data[basic_data['birthPlaceNormalized'].isin(us_birthplaces)]
foreign_born = basic_data[basic_data['birthPlaceNormalized'].isin(foreign_birthplaces)]
print('Average age of US-born Singhs', np.average(us_born['age']))
print('Average age of India-born Singhs', np.average(foreign_born['age']))
print('Youngest born in India', min(foreign_born['age']))
print('Oldest born in US', max(us_born['age']))
```
