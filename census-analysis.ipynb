```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import cluster, preprocessing
```

```python
basic_data = pd.read_json('1940_singh_census_basic.json')
basic_data.columns
```

```python
print('There were', len(basic_data.index), 'people in the United States 1940 Census with last name Singh')
```

```python
races = basic_data.groupby('raceRecordedNormalized').agg('count')
races = races.drop(races.columns[1:], axis = 1).rename(index=str, columns={'age': 'Count'})
```

```python
races
```

```python
race_pct = races['Count']/np.sum(races['Count'])
races['Percentage'] = race_pct
races
```
